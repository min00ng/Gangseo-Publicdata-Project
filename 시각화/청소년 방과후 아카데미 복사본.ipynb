## 2. 서울시 청소년 방과후 아카데미 시설

```python
import pandas as pd
import matplotlib as mp
```

```python
academy=pd.read_csv('서울시 청소년방과후아카데미 시설현황정보.csv')
academy
```

```python
from folium.map import Popup
import pandas as pd
import folium

df=pd.read_csv('서울시 청소년방과후아카데미 시설현황정보.csv')
seoul_map=folium.Map(location=[37.55,126.88],titles='Stamen Terrain',
                    zoom_start=13)
#위치 정보를 Marker 로 표시
#index를 시설명으로 변경하고 시설 이름이 표시되도록 함
for name, lat, lng in zip(df.시설명,df.위도,df.경도):
    iframe=folium.IFrame(name,width=100,height=80)
    popup=folium.Popup(iframe,max_width=500)
    folium.Marker([lat,lng],popup=popup).add_to(seoul_map)
    
seoul_map

#청소년 방과후 아카데미의 경우 강서구에는 1개만 입지
#다문화 가정 아이들을 위한 시설은 마련되어 있지 않음 또한 프로그램도 운영중이지 않음
```

## 3. 강서구 지역 아동센터

```python
center=pd.read_csv('지역아동센터.csv')
center
#지역 아동센터 서울시 지원단이 운영하는 강서구 지역 아동센터
#프로그램 운영에 대한 자세한 설명이 홈페이지에 나와있지 않고 홈페이지가 없는 경우가 대부분
```

```python
from folium.map import Popup
import pandas as pd
import folium

df=pd.read_csv('지역아동센터.csv')
seoul_map=folium.Map(location=[37.55,126.88],titles='Stamen Terrain',
                    zoom_start=13)
#위치 정보를 Marker 로 표시
#index를 시설명으로 변경하고 시설 이름이 표시되도록 함
for name, lat, lng in zip(df.시설명,df.위도,df.경도):
    iframe=folium.IFrame(name,width=100,height=80)
    popup=folium.Popup(iframe,max_width=500)
    folium.Marker([lat,lng],popup=popup).add_to(seoul_map)
    
seoul_map

#1~3학년 초등학생을 대상으로 하는 지역 아동 센터의 경우 19개 정도
#다문화 가정 아이들을 위한 프로그램은 마련되어 있지 않음
```

```python
#19개의 지역 아동 센터의 운영 상태 확인
#
#2030 비율이 높은 강서구의 특징에 맞추어 다문화 가정 학생들을 위한 멘토링 진행
#문화체험과 함께 학습 격차 해소에 도움

#논문인용 
#취학자녀가 있는 국제결혼 부부 대상 자녀의 학교생활 지원과 관련된 어려움을 조사한 결과 
#가장 큰 어려움으로서는 사교육비 부담(45.8%),8) 숙제 도와주기  (35.8%),9) 
#그 외 국제결혼 가정의 특수한 상황 속에서 의사소통이 안 된다든가 정체성 혼란 등에 대한 고민은 18.4%로 나타났다. (Cho  2007;  Lee,  2009)

#교육과학기술부는 교육현장에서의 다문화가정 학생 수의 증가에 따른 지원방안으로서 다문화가정 학생 멘토링 사업을 2009년부터 실시
#하지만 교대 중심으로 이루어져 멘토와 멘티 간 접근성 문제 발생
#따라서 강서구 내 2030을 다문화 가정 학생 멘토링과 연결
#지역 아동 센터 내 다문화 가정 아이들을 위한 돌봄, 멘토링 관련 프로그램을 추가하여 다문화 가정 아이들의 학교 이탈률을 막는 것이 목적
```
